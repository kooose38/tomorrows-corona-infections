```python
import pandas as pd 
import torch 
import torch.nn as nn 
import torch.nn.functional as F
from tqdm import tqdm 
import numpy as np 
from typing import Dict, List, Any, Tuple 
import os, time, pickle, json 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from statistics import mean 
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

device = "cuda:0" if torch.cuda.is_available() else "cpu"
print(device)
torch.manual_seed(0)
np.random.seed(0)
```

```python
class LoadDataset():
    def __init__(self, dataset):
        self.dataset = pd.read_csv(dataset).values
        self.train = []
        self.val = []
        self._split()

    def _split(self):
        n_span = 210 # 時期によって分布が大幅に変化すると予想されるので時期を特定してモデル作成する
        span = 30 
        self.dataset = self.dataset[:n_span, :]
        train, val = self.dataset[:-span, :], self.dataset[-span:, :]
        s = MinMaxScaler(feature_range=(0, 1))
        train_s = s.fit_transform(train)
        val_s = s.transform(val)

        for i in range(train.shape[0]-span):
            # 前１８０日分のデータを訓練用
            data = {}
            inputs = train_s[i:i+span]
            inputs = torch.FloatTensor(inputs) # (30, 6)
            target = train[i+span:i+span+1, 0].ravel()[0] # (1, )
            target = torch.tensor(target, dtype=torch.int64)

            data["inputs"] = inputs 
            data["target"] = target 
            self.train.append(data)

        for i in range(span):
            # 残り３０日分を検証用
            input_train = train_s[(-1)*span+i:, :] # 訓練データから直近３０日分のデータを取得
            input_val = val_s[:i, :] # 時系列順に１日ずつ追加する
            inputs = np.concatenate([input_train, input_val])
            target = val[i, 0].ravel()[0]
            data = {}
            data["inputs"] = torch.FloatTensor(inputs)
            data["target"] = torch.tensor(target, dtype=torch.int64)
            self.val.append(data)

data = LoadDataset("./train.csv")
```

```python
train_dl = torch.utils.data.DataLoader(data.train, batch_size=1, shuffle=False)
val_dl = torch.utils.data.DataLoader(data.val, batch_size=1, shuffle=False)
for r in train_dl:
    print(r["inputs"].size())
    print(r["inputs"])
    print(r["target"])
    break 
```

```python
class LSTM(nn.Module):
    def __init__(self, hidden_dim, input_size=6, tag_size=1):
        super(LSTM, self).__init__()
        self.lstm = nn.LSTM(input_size, hidden_dim, batch_first=True)
        self.fc = nn.Linear(hidden_dim, tag_size)
    
    def forward(self, x):
        # x.size (1, 30日分のデータ、各特長量6)
        hs, _ = self.lstm(x, None)
        out = hs[:, -1, :] # 時系列に従い最後の隠れ層を使う
        out = F.hardtanh(self.fc(out), 0, 80000)
        return out 


HIDDEN_DIM = 128
net = LSTM(HIDDEN_DIM)
```

```python
def train(net, x, t, criterion, optimizer):
    y = net(x)
    loss = criterion(y[0], t)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    return loss.item()

def save(net, e):
    os.makedirs("weights", exist_ok=True)
    torch.save(net.state_dict(), f"weights/net{str(e)}.pth")

def show_loss(train_loss, val_loss):
    plt.figure(figsize=(14, 5))
    plt.plot(np.arange(len(train_loss)).tolist(), train_loss)
    plt.plot(np.arange(len(val_loss)).tolist(), val_loss)
    plt.legend(["Train", "Val"])
    plt.grid()
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.show()

def trainer(net, train_dl, val_dl, num_epoch, lr=0.001):
    optimizer = torch.optim.Adam(net.parameters(), lr=lr)
    criterion = nn.L1Loss() # ２乗誤差平均
    net.to(device)
    plot_train, plot_val = [], []
    best_val_loss, best_net = np.inf, None 

    for e in range(num_epoch):
        net.train()
        e_train = []
        for i, data in enumerate(tqdm(train_dl)):
            x = data["inputs"].to(device)
            t = data["target"].to(device) 

            loss = train(net, x, t, criterion, optimizer)
            e_train.append(loss)

        plot_train.append(mean(e_train))

        net.eval()
        e_val = []
        for i, data in enumerate(tqdm(val_dl)):
            x = data["inputs"].to(device)
            t = data["target"].to(device)

            with torch.no_grad():
                y = net(x)
                loss = criterion(y[0], t).item()
            e_val.append(loss)
        plot_val.append(mean(e_val))

        if best_val_loss > mean(e_val):
            best_val_loss = mean(e_val)
            best_net = net 
    
    save(best_net, num_epoch)
    show_loss(plot_train, plot_val)
    del loss 
    torch.cuda.empty_cache()
    
    return best_net         

trained_net = trainer(net, train_dl, val_dl, 1000)
```

```python
trained_net_ = trainer(trained_net, train_dl, val_dl, 500) # ロスが減少傾向が続きそうなので追加で学習
```

```python
def load_net():
    HIDDEN_DIM = 128
    net = LSTM(HIDDEN_DIM)
    net.load_state_dict(torch.load("weights/net1500.pth", map_location={"cuda:0": "cpu"}))
    net.eval()
    return net

def show_eval(pred, corr):
    plt.figure(figsize=(15, 6))
    plt.plot(np.arange(len(pred)).tolist(), pred)
    plt.plot(np.arange(len(corr)).tolist(), corr)
    plt.legend(["predict", "correct"])
    plt.grid()
    plt.xlabel("days")
    plt.ylabel("Newly confirmed cases")
    plt.show()


def evaluate(val):
    net = load_net()
    predict, correct = [], []
    for data in tqdm(val):
        x = data["inputs"].to("cpu")
        t = data["target"].cpu().numpy().tolist()
        with torch.no_grad():
            y = net(x)
            y = y[0].cpu().numpy().tolist()
        predict.append(y[0])
        correct.append(t[0])

    loss = mean_squared_error(predict, correct)
    show_eval(predict, correct)


evaluate(val_dl)
# 滑らかな部分に対しては適合しているが異常な上昇には弱い
```

```python
def inference(test_csv):
    with torch.no_grad():
        s = MinMaxScaler(feature_range=(0, 1))
        test = pd.read_csv(test_csv).values
        test = s.fit_transform(test)
        test = test[-30:, :]
        test_tensor = torch.FloatTensor(test).unsqueeze(0)
        net = load_net()
        y = net(test_tensor)
        y = y[0].cpu().numpy().tolist()
        tomorrow = y[-1]
        print(f"明日の感染者数は{tomorrow:.1f}人と予想されます")

def transform(target_csv, hos_csv, death_csv, severe_csv):
    columns = ['Newly confirmed cases', 'Requiring inpatient care',
       'Discharged from hospital or released from treatment',
       'To be confirmed', 'Deaths(Cumulative)', 'Severe cases']
    target = pd.read_csv(target_csv)
    hos = pd.read_csv(hos_csv)
    death = pd.read_csv(death_csv)
    severe = pd.read_csv(sever_csv)

    target = target.groupby("Date").sum().reset_index()
    hos = hos.groupby("Date").sum().reset_index()
    death = death.groupby("Date").sum().reset_index()
    severe = severe.groupby("Date").sum().reset_index()
    hos_t = pd.merge(target, hos, how="outer", left_on="Date", right_on="Date")
    hos_t_d = pd.merge(hos_t, death, how="outer", left_on="Date", right_on="Date")
    df = pd.merge(hos_t_d, severe, how="outer", left_on="Date", right_on="Date")

    df = df.fillna(0)
    df["Date"] = pd.to_datetime(df.Date)
    df = df.groupby("Date").sum()
    df = df[columns]
    return df 

def main(target_csv, hos_csv, death_csv, severe_csv):
    # 厚生労働省のオープンデータセットを指定
    # https://www.mhlw.go.jp/stf/covid-19/open-data.html
    df = transform(target_csv, hos_csv, death_csv, severe_csv)
    inference(df)
    
# if __name__="__main__":
#     main()
```
